```python
not_null = lambda x : x.notnull()
def list_nan_values(df, function):
    check_nan = lambda x : len(df) - len(df[function(x)])
    list_nan_cols = list()
    for col in df.columns:
        list_nan_cols.append((col, check_nan(df[col])))
    return list_nan_cols
```

```python
list_nan = list_nan_values(df, not_null)
```

```python
import numpy as np
import math
def get_population(df, column):
    rs = {}
    values = list(column.dropna().unique())
    for val in values:
        rs[val] = len(df[column == val])
    rs[np.nan] = len(column.isnull()[lambda x: x == True])
    return rs
check_values = {}
for col in df.columns:
    check_values[col] = get_population(df,df[col])
check_values
```
