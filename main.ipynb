```python
def list_nan_values(df):
    check_nan = lambda x : len(df) - len(df[x.notnull()])
    list_nan_cols = list()
    for col in df.columns:
        list_nan_cols.append((col, check_nan(df[col])))
    return list_nan_cols
```

```python
list_nan = list_nan_values(df)
```

```python
for col, nan in list_nan:
    if nan >= 1200:
        df = df.drop(col, axis=1)
list_nan_values(df)
```
